```python
print(df["Major_Category"].value_counts())
print(df["Year_of_Study"].value_counts())
```

```python
plt.figure(figsize=(9,6))
sns.countplot(data=df,x="Major_Category",hue="Year_of_Study",width=0.5,palette="rainbow")
plt.xticks(rotation=0)
plt.title("Student Distribution by Major and Year of Study",fontsize=13,
          fontweight="bold")
plt.xlabel("Major Category", fontsize=13)
plt.ylabel("Number of Students", fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(title="Year_of_Study",bbox_to_anchor=(1,1),loc="upper left")
plt.tight_layout()
plt.savefig("Student Distribution.png",dpi=300,bbox_inches="tight")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.histplot(df,x="Weekly_GenAI_Hours",kde=True,palette="rainbow")
plt.title("Distribution of weekly GenAI Usage Hours",fontsize=13,
          fontweight="bold")
plt.xlabel("Weekly_GenAI_Hours", fontsize=13)
plt.ylabel("Number Of Students", fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.tight_layout()
plt.savefig("Distribution of weeekly GenAI usage Hours.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
use_case_counts=df["Primary_Use_Case"].value_counts()
colors = ['#4E79A7', '#F28E2B', '#59A14F', '#E15759', '#B07AA1']
plt.figure(figsize=(10,7))
plt.pie(use_case_counts,labels=None,colors=colors,autopct="%1.1f%%",startangle=90,explode=[0.02]*5,pctdistance=0.80,radius=1.15,wedgeprops=dict(width=0.42, edgecolor='white', linewidth=3),textprops={"fontsize":13})
plt.text(0, 0,"Generative AI\nUse Cases",ha='center',va='center',fontsize=22,fontweight='bold')
plt.title("Primary Use Cases of Generative AI",fontsize=24,fontweight='bold',pad=50)
legend_labels = ["Debugging","Copywriting","Ideation","Summarizing","Direct Answer"]
plt.legend(legend_labels,loc='upper center',bbox_to_anchor=(0.5, 1.10),ncol=3,frameon=False,fontsize=14)
plt.axis("equal")
plt.tight_layout()
plt.subplots_adjust(top=0.78)
plt.savefig("Primary Use Cases of Generative AI.png",dpi=300,bbox_inches="tight")
plt.show()
```
